## 1. Introduction

Cities are making more climate change policies, which is overall a good sign of progress. However, we should be critical of all policies and question 1) their effectiveness and 2) if the policy is creating unexpected consequences. For example, climate policies have the potential to both positively and negatively impact the economy and quality of life for residents. 

Living in London, I am interested in exploring the impact of the Ultra Low Emission Zone (ULEZ) on real estate values in the city. Anytime the government draws a boundary line and uses it to create spatial policy, it creates unintended externalities. The ULEZ was created and announced in April 2019  to "reduce air pollutant emissions from road transport" (Ma, et al, 2021). The ULEZ has been one of many policies that London has implemented with the aim to improve air quality in the capital.

There have been numerous academic studies to determine if the ULEZ and its relative, the LEZ have been effective as policy treatments for reducing urban emissions and improving air quality. Researchers have found there is a downward trend of air pollution over time and they did not find a significant impact by the ULEZ (Ma, et al, 2021)

Is it possible that the ULEZ has impacted real estate values and home prices in recent sales? Spatial policies have the ability to make the areas they effect in the city either more or less desirable to live in. This project will try to determine if house prices, as a proxy for desirability and demand of houses, have increased inside the zone. 

## 2. Background and Literature Review

Academics in the social sciences and public health often evaluate if a policy has been an effective intervention and made the difference it was created to do. We can evaluate if a policy is effective by using causal inference techniques. 

For spatial policies, we can consider the region and geography that the policy effects as the 'treatment' group, and the rest of the geography as a control group. This research question is inspired by a paper that compared house prices within school catchment districts of New York City (Rischard, et al., 2021), and that is how they framed their research. There are specific geographic boundaries that divide houses and streets into the districts, and given that schools fall on a ranking of desirability, the ability to live within the catchment area matters and house prices are impacted. 

This paper found that there is a significant difference in house prices across several districts. They found one area in New York City where houses on the desirable side of the district line were on averge roughly 20% more than the houses on the less desirable side using Geographical Regression Discontinuity Design (GeoRDD). Of course, housing stock is hetergenious and one must acknowledge the differences and spatial variations of housing type and quality, no to mention community ammenities such as schools and high streets, that impact prices. 

Other researchers have dived into spatial policy in London specifically. Percoco studied the impact of the original Congestion Charge in 2003. The Congestion Charge covered the very central area of London and was the same area as the ULEZ in 2019. Percoco found unclear effects of the Congestion Charge on air pollution. After ULEZ was introduced, Ma, Graham, and Stettler (2021) used Regression Discontinuity Design to determine if the ULEZ has impacted air quality. They found no statistically significant impact at the time of the policy in 2019, so they include the policy was not effective on its own. 

We can take these approaches and apply them to house prices to see if these policy zones have impacted the real estate in central London to make them more or less desirable. 




## 3. Research Question

What is the impact of the ULEZ on London House Prices?

We set up the Null Hypothesis that there is no difference between the house prices inside the ULEZ and outside the ULEZ.

```python
#Import libraries 
import pandas as pd # version 2.1.4
import os
import seaborn as sns #version 0.13.1
import glob
import matplotlib.pyplot as plt
import numpy as np #verion 1.26.3
import datetime

import plotly
import plotly.express as px
import warnings
from statsmodels.formula.api import ols
from statsmodels.iolib.summary2 import summary_col
import matplotlib.pyplot as plt


warnings.filterwarnings('ignore')
sns.set(font_scale=1.5)
sns.set_style("white")
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
#libraries for spatial data
import geopandas as gpd
import geoplot
import geoplot.crs as gcrs
```

### Dataset 1: Create a dataset of house prices by type and year

```python
#load in house prices dataset 

def my_csv(filepath):
    """Define my own function to read in the year data"""
    year = filepath[-8:-4]
    names = ["postcode", "detached_" + year,  "semi-det_" + year, "terraced_" + year, 
          "Flat/mais_" + year, "average_" + year ]
   # print(names)
    #names = ['A', 'B', 'C', 'D', 'E', 'F']
    d = pd.read_csv(filepath, names = names, skiprows = 6, usecols = [0, 1, 3, 5, 7, 9])
    return  d.set_index('postcode')

house_prices = pd.concat(map(my_csv, glob.glob('Data/avgPrice*.csv')), axis = 1 , join='inner')
```

```python
house_prices.head()
```

```python
house_prices.to_csv("house_prices_all_years.csv")
```

```python
## add in columns for the year values 
house_prices.insert(0, 'year_2015', 2015, True)
house_prices.insert(6, 'year_2016', 2016, True)
house_prices.insert(12, 'year_2017', 2017, True)
house_prices.insert(18, 'year_2018', 2018, True)
house_prices.insert(24, 'year_2019', 2019, True)
house_prices.insert(30, 'year_2020', 2020, True)
house_prices.insert(36, 'year_2021', 2021, True)
house_prices.insert(42, 'year_2022', 2022, True)
house_prices.insert(48, 'year_2023', 2023, True)


year_totals = house_prices[['average_2015', 'year_2015', 'average_2016', 'year_2016',
                            'average_2017', 'year_2017', 'average_2018', 'year_2018',
                            'average_2019', 'year_2019','average_2020', 'year_2020',
                           'average_2021','year_2021', 'average_2022', 'year_2022',
                            'average_2023', 'year_2023']]
```

### Dataset 2: Now load in the ULEZ Geometry

```python
#load in the ULEZ geometry from the downloaded shapefile and set the CRS
ulez_geo = gpd.read_file('Data/ULEZShapefile/UltraLowEmissionsZoneBoundary(ULEZ).shp')
ulez_geo = ulez_geo.to_crs('epsg:4326')
```

```python
#print out the CRS and confirm it is correct
ulez_geo.crs
```

### Dataset 3: Import the postcode sector geometries 

```python
#create a list of London Postcodes, and then split to find the first half of the postcode
#the Geojson files for the geometries are labeled by the first half of the postcode, ie 'SW4'
copy = year_totals.reset_index()
london_postcodes = copy['postcode'].tolist()
london_folder = [p.split(' ')[0] for p in london_postcodes]
london_folder = np.unique(london_folder)
```

```python
# import geojsons from folders with London postcode sectors
json_dir_name = "Data/gb-postcodes-v5/sectors"
list_of_london_folders = london_folder 

#create  a list of folder paths to look for geojson files
list_of_london_folder_paths = [json_dir_name + '/' + f for f in list_of_london_folders]

file_list_london = []
for f in list_of_london_folder_paths:     
    json_pattern = os.path.join(f,'*.geojson')
    file_list = glob.glob(json_pattern)
    file_list_london.extend(file_list)

```

```python
#read in all geojson files and create one dataframe
list_df = []
for i in file_list_london: 
    df = gpd.read_file(i)
    list_df.append(df)

london_geo = pd.concat(list_df)

#confirm the correct CRS - EPSG 4326
london_geo.crs
```

### Data Exploration: Visualize the data 

Check what the postcode sectors look like on a map. The London postcode sectors are white and black out line and the ULEZ is in Red, in the very center of the city.


```python
#map the geometries together to visualize
base = london_geo.plot(color='white', edgecolor='black')
ulez_geo.plot(ax=base, marker='o', color='red', markersize=5, alpha= 0.5)
```

Next, find the group of postcode sectors that intersect with the boundary of the ULEZ

```python
#use the intersects() function to get the postcode geometries that intersect with the boundary
intersection = london_geo[london_geo.intersects(ulez_geo.boundary)]

#use the within() function to get the postcode geometries that are completely within the ULEZ boundary
within = london_geo[london_geo.within(ulez_geo)]
```

Below, there is a map of the postcode sectors in white and blue. White post code sectors are the ones that intersect with the ULEZ and the blue ones are completely within the ULEZ. A limitation of this methodolgy is that it was difficult to get the polygons separated along the line of the ULEZ. Ideally, the postcode sectors would be on either side of the ULEZ, instead of straddling the boundary. For this reason, this causes an issue in the experiment design. 

```python
#map the two groups of postcode sectors with the ULEZ 
intersection_base = intersection.plot(color='white', edgecolor='black')

within.plot(ax = intersection_base, color = 'blue', edgecolor = 'black')
ulez_geo.plot(ax=intersection_base, marker='o', color='red', markersize=5, alpha= 0.5)
```

### Create Control and Treatment groups

Based on the visualization above, the postcode sectors can be broken up into a control group and a treatment group. In this case, because the sectors cannot be clearly within or without, we will define the control group as any sector intersecting the ULEZ boundary, and the treatment group are those sectors completely within the ULEZ.

```python
#create a list of postcode sectors in control group
postcode_sector_control = intersection['sector'].tolist()
```

```python
#create a list of postcode sectors in treatment group
postcode_sector_treat = within['sector'].tolist()
```

```python
#create a copy  of the house price data and reset index
copy = year_totals.reset_index()
```

```python
#create a control dataset from the postcodes in the control group and a make a long version for visualizing
mask_control = copy['postcode'].isin(postcode_sector_control)
control = copy[mask_control]
control.set_index('postcode')
```

```python
control_long = pd.wide_to_long(control, ['average_', 'year_'], i = 'postcode', j = 'year')
```

```python
control_long.head()
```

```python
#create a treatment dataset from  the postcodes in the treatment group  and make  a long version for visualization
mask_treat = copy['postcode'].isin(postcode_sector_treat)
treatment = copy[mask_treat]
treatment.set_index('postcode')
```

```python
treatment_long = pd.wide_to_long(treatment, ['average_', 'year_'], i = 'postcode', j = 'year')
```

```python
#Plot the control and treatment groups
plt.plot(control_long['year_'], control_long['average_'], 'o')
plt.plot(treatment_long['year_'], treatment_long['average_'], 'o')
plt.show()
```

Above we have the control (orange) and treatment (blue) groups shown over time. This is not the most helpful graph because the price values are collected at one date. We can see that there is some variation and outliers of price averages in both groups.  

## Difference in Differences

Below the Difference in Differences technique is applied to the treatment and control groups. First, we need to label the control groups as not treatment, and set the treatment value to zero. The treatment group is labeled as 1, and then we combine the datasets into one dataset. 

```python
control['treatment'] = 0
treatment['treatment'] = 1
did = pd.concat([control, treatment] )
```

```python
did
```

```python
# we then want to pivot the dataset to long so we have each postcode and year next to the average. 
did.set_index('postcode')
did = pd.wide_to_long(did, ['average_', 'year_'], i = 'postcode', j = 'year')
```

```python
#add a column for the datetime so we can graph it over time 
did['date'] = did.apply(lambda row: datetime.datetime(row['year_'], 1, 1), axis = 1)
```

```python
# Create a variable and assign the datapoints that are 2019 and afterward to post group
did['post']=np.where(did['date']>='2019-01-01',1,0)
```

```python
# create a variable that is 1 if the date is after the ULEZ policy and the treatment group and 0 otherwise
did['post_treatment']=did['post']*did['treatment'] 
```

```python
did.head()
```

Visualize the treatment and control group overtime so we can see how they vary

```python
#now graph the values to visualize
sns.lineplot(data=did,x='date',y='average_',hue='treatment')
plt.axvline(pd.to_datetime('2019-01-01'),color='black',linestyle='dashed', label='ULEZ Policy Enacted')
plt.title('Average house prices over time')
plt.legend(['Control Average', 'Control Range', 'Treatment Average', 'Treatment Range'])
```

Visualize the treatment and control as box plots so we can see how they vary with pre and post policy.

```python

#create a category for the plot below 
did['category']=did['treatment'].astype(str)+did['post'].astype(str) 

# Define colour palette
palette = {0: "blue", 1: "orange"}

sns.boxplot(x='category', y='average_', hue='treatment', data=did, palette=palette).set_xticklabels(["Pre x Control", "Pre x Treatment",'Post x Control','Post x Treatment']) 

# Get the current legend
leg = plt.legend()

# Define new labels for the legend
new_labels = ['0 (Control)', '1 (Treatment)']

# Set the new labels
for t, label in zip(leg.texts, new_labels): 
    t.set_text(label)

plt.xlabel('')
plt.title('House Sale Prices by Treatment and Post Treatment')
plt.show()
```

Run the test to see if this is statistically significant using OLS model 

```python
did_model = ols('average_ ~  post + treatment + post_treatment', did).fit()
print(did_model.summary())
```

## Results

From the results of the Difference in Differences summary table we can see the following variables: 

post: is positive and not statistically significant at the 5% level
treatment: is positive and not statistically significant 
post_treatment: is positive and not statistically significant. As our differnce-in-differences estimator, because it is statistically insignificant, we cannot prove our null hypothesis wrong. It presents such that there is no difference in the treatment and control group after the 2019 policy change.  


From the visualizations above, we can see there is a lot of variation in treatment group values. There is a slight upward trend before the 2019 cutoff for the treatment, and there is not a similar trend in the control group. Therefore, the test does not really establish that the groups are similar. Then, after 2019, the treatment group prices fluctuate a lot and have a dip in 2020, which we can attribute to COVID. 

# Conclusion and Discussion

I have compared the control and treatment groups using the Difference in Difference test. 
We cannot conclude that our null hypothesis is false because the post-treatment variable had no statistically significance on the OLS model. This means we cannot conclude that the ULEZ policy had any impact on house prices before and after 2019. 

We can attribute some error in this model to the imperfect separation of the geometries. The nature of the ULEZ boundary made it difficult to separate postcode sectors into control and treatment groups based on geometries. Instead, I would consider finding another geometry to break up the groups, however then we would not have the house price data at that scale. 

Perhaps there is also an issue with real estate property speculation, because this is a policy that was discussed before it was enacted, so purchasers might have acted anticipating this policy. That would mean the 2019 boundary would be weaker in this analysis.
